```python
df_clean = df.copy()
```

```python
df_clean['gender'] = df_clean['gender'].map({
    'male':0, 
    'female':1
})
```

```python
df_clean['test preparation course'] = df_clean['test preparation course'].map({
    'none':0,
    'completed':1
})
```

```python
df_clean['lunch'] = df_clean['lunch'].map({
    'standard':0,
    'free/reduced':1
})
```

```python
df_clean = pd.get_dummies(df_clean,columns=['race/ethnicity'], drop_first=True)
df_clean = pd.get_dummies(df_clean,columns=['parental level of education'], drop_first=True)
```

```python
x = df_clean.drop(["math score", "reading score", "writing score"],axis=1)
y = df_clean['avg_score']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
cat_cols
```

```python
cat_cols_x = x.select_dtypes(include=['object']).columns
num_cols_x = x.select_dtypes(exclude=['object']).columns
```

preprocessing

```python
preprocessor = ColumnTransformer([
    ('num','passthrough',num_cols_x),
    ('cat',OneHotEncoder(drop='first', handle_unknown='ignore'),cat_cols_x)
])
```

models

```python
models = {
    'LinearRegression': LinearRegression(),
    'DecisionTreeRegressor': DecisionTreeRegressor(random_state=42), 
    'RandomForestRegressor': RandomForestRegressor(n_estimators=100,random_state=42), 
    'GradientBoostingRegressor': GradientBoostingRegressor(random_state=42), 
    'SVR': SVR(kernel='rbf')
}
```

```python
x_train , x_test , y_train ,y_test = train_test_split(x , y , test_size=0.2, random_state=42)
```

```python
results = []

for name , model in models.items(): 
    pipe = Pipeline([
        ('preprocessor',preprocessor), 
        ('scaler', StandardScaler() if name == 'SVR' else 'passthrough'),
        ('model', model)
    ])
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)

    # mean square for comp and r2 for better visualize comp marices for regression accuracy for classification
    mse = mean_squared_error(y_test,y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test,y_pred)

    results.append({
        'Model':name, 
        'mean square error': mse, 
        'root mean square error': rmse, 
        'R2 score': r2
    })

results_df = pd.DataFrame(results).sort_values(by= 'R2 score',ascending= False)
print(results_df)

# bar plot of r2 
plt.figure(figsize=(16,9))
sns.barplot(data=results_df, x= 'Model', y = 'R2 score', palette='viridis')
plt.title('Model comparison of RÂ² score')
plt.show()
```
